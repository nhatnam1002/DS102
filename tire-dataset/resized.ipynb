```python
import glob2
import math  
import numpy as np
import cv2
import os
def ConvertImage(image_name): #tao ham resized
  print(image_name+" process")
  img = cv2.imread(image_name) 
  img=cv2.resize(img, (224, 224))
  cv2.imwrite("resized_"+image_name[:-3]+"jpg",img)
  print(image_name+" done")
```

```python
%cd /content/gdrive/MyDrive/DS102/tire-dataset/Dataset/test/
files = []
for ext in ["*.png", "*.jpeg", "*.jpg","*.jfif", "*.heic"]:
  image_files = glob2.glob(os.path.join( ext))
  files += image_files
print(files)
for idx in np.arange(len(files)):ConvertImage(files[idx])
```
